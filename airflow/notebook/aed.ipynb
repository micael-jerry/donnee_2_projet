```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://donnee-2-bucket.s3.eu-west-3.amazonaws.com/data.csv"
data:pd.DataFrame = pd.read_csv(url)
```

```python
# Afficher les premières lignes du DataFrame pour avoir un aperçu des données
data.head()
```

```python
# Vérifier les informations sur les types de données et les valeurs manquantes
data.info()
```

```python
# Statistiques descriptives
data.describe()
```

```python
# Visualiser la distribution des valeurs par date
date_columns = data.columns[3:]
data[date_columns].plot(kind='box', vert=False, figsize=(12, 6))
plt.title('Distribution des valeurs par date')
plt.xlabel('Valeur')
plt.show()
```

```python
# Visualiser l'évolution temporelle pour quelques symboles
symbols_to_plot = data['symbols'].sample(n=5, random_state=42)
for symbol in symbols_to_plot:
    symbol_data = data[data['symbols'] == symbol]
    plt.plot(symbol_data[date_columns].T, label=symbol)

plt.title("Évolution temporelle de quelques symboles")
plt.xlabel("Date")
plt.ylabel("Valeur")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualiser la corrélation entre les symboles et les pays
correlation_matrix = data[date_columns].corr()
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.title('Matrice de corrélation entre les symboles')
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.show()
```
